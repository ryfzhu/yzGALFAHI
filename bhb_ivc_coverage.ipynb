To check where Jess's BHB stars are as compared to the IVCs on the sky. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from astropy.table import Table
```

```python
# qso = Table.read('myqso.txt', format='ascii')
# bhb = Table.read('werkbhb_lb.txt', format='ascii')
bhb = Table.read('BHBs/Hannah19_BHBs_l_b.txt', format='ascii')
bhb_l = np.array([float(l) for l in bhb['l']])
bhb_b = np.array([float(b) for b in bhb['b']])
```

```python
fig = plt.figure(figsize=(8, 4))
ax = fig.add_subplot(111, projection="mollweide")


# the QSOs on the sky
#x_qso = np.radians(180-qso['col2'])
#y_qso = np.radians(qso['col3'])
#plt.scatter(x_qso, y_qso, label='QSOs')

# some IVCs from Table 1 on page 37 in the HVC book by van Woerden 2004
x_ivarch = np.radians(180-np.array([220, 220, 80, 80, 220, ]))
y_ivarch = np.radians(np.array([85, 30, 30, 85, 85]))
plt.plot(x_ivarch, y_ivarch, color='r', label='IV Arch')

x_ivspur = np.radians(180-np.array([290, 290, 210, 210, 290]))
y_ivspur = np.radians(np.array([90, 45, 45, 90, 90]))
plt.plot(x_ivspur, y_ivspur, color='g', label='IV Spur')

x_llivarch = np.radians(180-np.array([185, 185, 110, 110, 185]))
y_llivarch = np.radians(np.array([37, 27, 27, 37, 37]))
plt.plot(x_llivarch, y_llivarch, color='y', label='LLIV Arch')

x_oa = np.radians(180-np.array([185, 185, 54, 54, 185]))
y_oa = np.radians(np.array([31, -5, -5, 31, 31]))
plt.plot(x_oa, y_oa, color='c', label='Outer Arm')

#x_complexk = np.radians(180-np.array([70, 70, 30, 30, 70]))
#y_complexk = np.radians(np.array([68, 25, 25, 68, 68]))
#plt.plot(x_complexk, y_complexk, color='b', label='Complex K')

x_acshell = np.radians(180-np.array([200, 200, 160, 160, 200]))
y_acshell = np.radians(np.array([20, -20, -20, 20, 20]))
plt.plot(x_acshell, y_acshell, color=plt.cm.Oranges(0.5), label='AC Shell')

#x_pparch = np.radians(180-np.array([170, 170, 83, 83, 170]))
#y_pparch = np.radians(np.array([-33, -78, -78, -33, -33]))
#plt.plot(x_pparch, y_pparch, color=plt.cm.Oranges(0.5), label='PP Arch')

# the BHB on the sky
x_bhb = np.radians(180-bhb_l)
y_bhb = np.radians(bhb_b)
plt.scatter(x_bhb, y_bhb, label='BHBs', marker='*', s=70, color='k')

plt.legend(fontsize=10, loc='lower left')

# plot layout 
plt.grid(linestyle='--', alpha=0.5)
ax.set_xticklabels(['', '', r'$270\degree$', '', '', r'$180\degree$', '', '', 
                    r'$90\degree$', '', ''], fontsize=12, alpha=0.6)
ax.set_yticklabels(['', r'$-60\degree$', '', r'$-30\degree$', '', r'$0\degree$', 
                    '', r'$30\degree$', '', r'$60\degree$', ''], fontsize=12, alpha=0.6)

plt.savefig('hannah19_bhbivc_center180.pdf')
```

```python
fig = plt.figure(figsize=(7, 4))
ax = fig.add_subplot(111, projection="mollweide")

al = 0.7
# some IVCs from Table 1 on page 37 in the HVC book by van Woerden 2004
x_ivarch_1 = np.radians(np.array([180, 180, 80, 80, 180]))
y_ivarch_1 = np.radians(np.array([85, 30, 30, 85, 85]))
x_ivarch_2 = np.radians(np.array([220, 220, 180, 180, 220])-360)
y_ivarch_2 = np.radians(np.array([85, 30, 30, 85, 85]))
plt.plot(x_ivarch_1, y_ivarch_1, color='r', label='IV Arch', lw=1, alpha=al)
plt.plot(x_ivarch_2, y_ivarch_2, color='r', label=None, lw=1, alpha=al)

x_ivspur = np.radians(np.array([290, 290, 210, 210, 290])-360)
y_ivspur = np.radians(np.array([90, 45, 45, 90, 90]))
plt.plot(x_ivspur, y_ivspur, color='g', label='IV Spur', lw=1, alpha=al)

x_llivarch_1 = np.radians(np.array([180, 180, 110, 110, 180]))
y_llivarch_1 = np.radians(np.array([37, 27, 27, 37, 37]))
x_llivarch_2 = np.radians(np.array([185, 185, 180, 180, 185])-360)
y_llivarch_2 = np.radians(np.array([37, 27, 27, 37, 37]))
plt.plot(x_llivarch_1, y_llivarch_1, color='y', label='LLIV Arch', lw=1, alpha=al)
plt.plot(x_llivarch_2, y_llivarch_2, color='y', label=None, lw=1, alpha=al)

x_oa_1 = np.radians(np.array([180, 180, 54, 54, 180]))
y_oa_1 = np.radians(np.array([31, -5, -5, 31, 31]))
x_oa_2 = np.radians(np.array([185, 185, 180, 180, 185])-360)
y_oa_2 = np.radians(np.array([31, -5, -5, 31, 31]))
plt.plot(x_oa_1, y_oa_1, color='c', label='Outer Arm', lw=1, alpha=al)
plt.plot(x_oa_2, y_oa_2, color='c', label=None, lw=1, alpha=al)

#x_complexk = np.radians(np.array([70, 70, 30, 30, 70]))
#y_complexk = np.radians(np.array([68, 25, 25, 68, 68]))
#plt.plot(x_complexk, y_complexk, color='b', label='Complex K')

x_acshell_1 = np.radians(np.array([180, 180, 160, 160, 180]))
y_acshell_1 = np.radians(np.array([20, -20, -20, 20, 20]))
x_acshell_2 = np.radians(np.array([200, 200, 180, 180, 200])-360)
y_acshell_2 = np.radians(np.array([20, -20, -20, 20, 20]))
plt.plot(x_acshell_1, y_acshell_1, color=plt.cm.Oranges(0.5), label='AC Shell', lw=1, alpha=al)
plt.plot(x_acshell_2, y_acshell_2, color=plt.cm.Oranges(0.5), label=None, lw=1, alpha=al)

# the BHB on the sky
bhb_l[bhb_l>180] = bhb_l[bhb_l>180]-360
x_bhb = np.radians(bhb_l)
y_bhb = np.radians(bhb_b)
plt.scatter(x_bhb, y_bhb, label='BHBs', marker='o', facecolor='none', s=20, edgecolor='k', lw=1)

#x_pparch = np.radians(180-np.array([170, 170, 83, 83, 170]))
#y_pparch = np.radians(np.array([-33, -78, -78, -33, -33]))
#plt.plot(x_pparch, y_pparch, color=plt.cm.Oranges(0.5), label='PP Arch')
plt.legend(fontsize=10, loc='lower center', bbox_to_anchor=(0.35, 0.25))

#plt.xlim(-np.pi, np.pi)
#plt.ylim(-np.pi/3, np.pi)

# plot layout 
plt.grid(linestyle='--', alpha=0.5)
ax.set_xticklabels([r'$-150\degree$', r'$-120\degree$', '', '', '', r'$0\degree$', 
                    r'$30\degree$', r'$60\degree$', r'$90\degree$', 
                    r'$120\degree$', r'$150\degree$'], fontsize=10, alpha=0.6)
ax.set_yticklabels(['', r'$-60\degree$', '', r'$-30\degree$', '', r'$0\degree$', 
                    '', r'$30\degree$', '', r'$60\degree$', ''], fontsize=10, alpha=0.6)

plt.savefig('BHBs/hannah19_bhbivc_center0.pdf')
```
